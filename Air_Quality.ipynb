## **AIR QUALITY ANALYSIS**

# **Introduction**

Air pollution has become a critical issue globally, with severe health and environmental implications. India, in particular, faces significant challenges, with air pollution levels often surpassing safe limits. This report focuses on analyzing air quality data from various states in India to explore trends, identify key pollutants, and visualize the impact across regions using Tableau. By leveraging data from the Ministry of Environment and Forests and Central Pollution Control Board of India, we can gain insights into the air quality situation and its progression over time.

# ***Data Overview***

Welcome to the Air Quality Analysis project! Air pollution is a critical environmental issue that poses significant health risks globally. According to the World Health Organization (WHO), air pollution is linked to 7 million premature deaths annually, making it the world's largest single environmental risk. In India, the situation is particularly alarming, with air pollution levels that rival and, in some cases, surpass those of China. This report details the development and use of a Tableau dashboard to analyze and visualize air quality data across different states in India, offering insights into the country's pollution levels and trends.

# **Project Statement**

The project aims to create an interactive Tableau dashboard that visualizes historical air quality data across various states in India. The dashboard is designed to help users explore the data, identify trends, and draw comparisons between regions. By leveraging this tool, stakeholders can gain a deeper understanding of air pollution in India and make informed decisions based on the insights derived from the data.

# **Objectives**

- Visualize Air Quality Across Regions: Provide a geographical representation of
  air quality levels across different states in India.
- Analyze Trends Over Time: Use time series charts to track changes in air    pollution levels over the years.
- Enable Comparative Analysis: Allow users to compare air quality between different regions or states.
- Identify Key Pollutants: Highlight the major pollutants contributing to poor air quality in various regions.
- Enhance User Interaction: Implement filters and annotations to enable customized analysis and provide context to the data.

# **Outcomes**

- Enhanced Understanding of Regional Air Quality: The dashboard offers a clear visual representation of air quality across Indian states, helping users identify the most and least polluted regions.
- Insightful Trends and Patterns: Time series analysis enables the identification of trends and patterns in air quality over time, offering insights into the effectiveness of pollution control measures.
- Comparative Insights: Users can easily compare air quality across different regions, facilitating a deeper understanding of regional disparities in pollution levels.
- Focused Pollutant Analysis: The dashboard highlights the top pollutants in each region, allowing for targeted discussions and policy recommendations.
- Interactive Exploration: With filters and annotations, users can explore the data according to their specific interests, making the analysis more personalized and relevant.

# **Features**

Data collection and Preparation

   1. Gathered the data from the sources.
   2. Cleaned and structured the data for analysis.


Map Visualization:

1. Displays air quality levels across India using geographical maps.
2. States and regions are color-coded based on the Air Quality Index (AQI), providing an immediate visual cue of pollution levels.


Time Series Charts:

1. Shows changes in air quality over the years through line charts.
2. Data can be grouped by month or year to reveal long-term trends and seasonal variations.

Filter Options:

1. Users can apply filters to select specific states, periods, or types of pollutants, enabling a focused analysis.
2. Customizable views allow for a more detailed exploration of the data.

Comparative Analysis:

1. Bar and line charts enable side-by-side comparisons of air quality metrics between different states or regions.
2. Highlights the variation in pollution levels, aiding in the identification of outliers and trends.

Top Pollutant Analysis:

1. Bar and pie charts display the concentration and proportion of various pollutants.
2. Helps in identifying the primary contributors to air quality degradation in specific areas.

# **How to Use**

- Data Import and Preparation:

    1. Begin by importing the cleaned air quality dataset into Tableau.
    2. Ensure that geographical fields (e.g., states) are correctly mapped for accurate visualization.

- Creating Visualizations:

    1. Start by building basic visualizations such as bar charts and line charts to get a feel for the data.
    2. Use Tableau's map feature to create geographical visualizations, ensuring that color coding is applied based on AQI levels.

- Adding Interactivity:

    1. Implement filters to allow users to narrow down the data based on states, time periods, or pollutants.
    2. Use actions to link different visualizations, creating an interactive experience where changes in one chart reflect in others.

- Publishing the Dashboard:

    1. Once the dashboard is complete, publish it on Tableau Public or Tableau Server for sharing.
    2. Ensure that the dashboard is user-friendly, with clear labels, legends, and instructions for use.

# **Conclusion**

The Tableau dashboard created for analyzing India's air quality data serves as a powerful tool for visualizing, exploring, and understanding the country's pollution levels. Through the use of interactive maps, time series charts, and comparative analyses, users can gain deep insights into regional air quality trends, the impact of pollutants, and potential areas for intervention. This tool not only aids in research and policymaking but also empowers the general public with valuable information about the air they breathe.
